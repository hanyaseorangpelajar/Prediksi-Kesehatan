# Data Preparation

```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module="sklearn.base")
```

```python
file = "/content/Sleep_health_and_lifestyle_dataset.csv"
data = pd.read_csv(file)
data.info()
data.head()
```

#

```python
# Menghapus data kolom yang tidak digunakan
data.drop(['Person ID', 'Gender', 'Sleep Disorder'] ,axis=1, inplace=True)
data.head()
```

# Data Splitting

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Label Encoding
label_encoder = LabelEncoder()
# Kolom 'Occupation'
data['Occupation'] = label_encoder.fit_transform(data['Occupation'])
label_mapping_occupation = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

# Kolom 'BMI Category'
data['BMI Category'] = label_encoder.fit_transform(data['BMI Category'])
label_mapping_bmi = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

# Kolom 'Blood Pressure'
data['Blood Pressure'] = data['Blood Pressure'].apply(lambda x: int(x.split('/')[0]))

print("Label Mapping for Occupation:", label_mapping_occupation)
print("Label Mapping for BMI Category:", label_mapping_bmi)
```

```python
# Membagi data menjadi atribut dan label
X = data[['Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'BMI Category', 'Blood Pressure', 'Heart Rate', 'Daily Steps']]
y = data['Stress Level']

# Membagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

# Model Building

```python
from sklearn.tree import DecisionTreeClassifier

# membuat model Decision Tree
tree_model = DecisionTreeClassifier()

# Melatih model dengan menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)
```

```python
# Evaluasi Model
from sklearn.metrics import accuracy_score

y_pred = tree_model.predict(X_test)

acc_secore = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_secore)
```
